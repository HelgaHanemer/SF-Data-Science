### <center>**НАЖБП**</center>
НАЖБП – это состояние, описываемое как избыточное накопление жира в форме триглицеридов (стеатоз) в печени (гистологически в > 5% гепатоцитов). У подгруппы пациентов с НАЖБП имеется повреждение клеток печени и воспаление в дополнению к избыточному жиру (стеатогепатит). Последнее состояние, носящее название НАСГ, буквально неотличимо гистологически от алкогольного стеатогепатита (АСГ). В то время как простой стеатоз, наблюдаемый при НАЖБП не коррелирует с ростом кратковременной заболеваемости и смертности, прогрессирование заболевания до НАСГ резко увеличивает риск развития цирроза, печеночной недостаточности и гепатоцеллюлярной карциномы (ГЦК). Цирроз, вызванный НАСГ – это более частая причина для трансплантации печени. Так как заболеваемость и смертность пациентов с НАСГ значительно возрастают, они еще сильнее коррелируют заболеваемостью и смертностью от сердечно-сосудистых заболеваний.

**Факторы риска развития НАЖБП**

+ Несбалансированное питание, употребление жирной, жареной пищи, сладкого, злоупотребление кофеином, никотином, алкоголем
+ Хронический стресс
+ Ожирение, высокий уровень холестерина и триглицеридов в крови
+ Генетические факторы
+ Сопутствующая патология (сахарный диабет II типа, заболевания сердечно-сосудистой системы, гормональные и обменные нарушения).

*по данным Всемироной Гастроэнтерологической организации,  ["Практические рекомендации Всемирной Гастроэнтерологической Организации"](https://www.worldgastroenterology.org/guidelines/nafld-nash/nafld-nash-russian)*

Важной частью любого исследования является расчет корреляции. Сначала напишем функцию, а затем построим тепловую карту.

```python
# Напишем функцию для интерпретации результата коэффициента корреляции

def correlation_tie(correlation: float) -> str:
    """
    Определяет уровень корреляции на основе переданного значения корреляции.

    Параметры:
    correlation (float): Значение корреляции, которое должно находиться в диапазоне от 0 до 1.

    Возвращает:
    str или np.nan: Описание уровня корреляции:
        - "Практически нет корреляции" для значений от 0 до 0.2
        - "Слабая корреляция" для значений от 0.2 до 0.5
        - "Умеренная корреляция" для значений от 0.5 до 0.7
        - "Сильная корреляция" для значений от 0.7 до 0.9
        - "Очень сильная корреляция" для значений от 0.9 до 1
        - np.nan для некорректных значений корреляции (вне диапазона 0-1)
    """
    if isinstance(correlation, (int, float)) and 0 <= correlation < 0.2:
        return "Практически нет корреляции"
    elif isinstance(correlation, (int, float)) and 0.2 <= correlation < 0.5:
        return "Слабая корреляция"
    elif isinstance(correlation, (int, float)) and 0.5 <= correlation < 0.7:
        return "Умеренная корреляция"
    elif isinstance(correlation, (int, float)) and 0.7 <= correlation < 0.9:
        return "Сильная корреляция"
    elif isinstance(correlation, (int, float)) and 0.9 <= correlation <= 1:
        return "Очень сильная корреляция"
    else:
        return np.nan
# посмотрим отдельно корреляцию между весом и исходом заболевания
correlationIO = nafld1_data['weight'].corr(nafld1_data['bmi'])
# Применение функции к результату корреляции
result = correlation_tie(correlationIO)
# Вывод результата
print(f"Корреляция между 'weight' и 'bmi': {correlationIO}")
print(f"Уровень корреляции: {result}")

# посмотрим отдельно корреляцию между весом и исходом заболевания
correlationIO = nafld1_data['male'].corr(nafld1_data['status'])
# Применение функции к результату корреляции
result = correlation_tie(correlationIO)
# Вывод результата
print(f"Корреляция между 'male' и 'status': {correlationIO}")
print(f"Уровень корреляции: {result}")
```

### **<center> Метод межквартильного размаха (метод Тьюки) </center>**

```python
#Построим гистограмму и коробчатую диаграмму для bmi:

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
histplot = sns.histplot(data=combine_data, x='bmi', ax=axes[0]);
histplot.set_title('bmi');
boxplot = sns.boxplot(data=combine_data, x='bmi', ax=axes[1]);
boxplot.set_title('bmi');
```

```python
import pandas as pd
from typing import Tuple

def outliers_iqr(data: pd.DataFrame, feature: str) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Выявляет выбросы в заданном признаке DataFrame с использованием метода межквартильного размаха (IQR)
    и возвращает DataFrame с потенциальными выбросами и очищенный DataFrame без них.
    
    Функция принимает на вход DataFrame и название признака, по которому необходимо определить выбросы,
    а затем возвращает два значения: DataFrame с потенциальными выбросами и очищенный DataFrame без них.
    
    Метод Тьюки использует 1-й и 3-й квартили для определения границ выбросов, которые рассчитываются как:
    - Нижняя граница: 1-й квартиль - (IQR * 0.8)
    - Верхняя граница: 3-й квартиль + (IQR * 0.8)

    Параметры:
    ----------
    data : pandas.DataFrame
        Входной DataFrame, содержащий данные для анализа.
        
    feature : str
        Название признака (колонки) в DataFrame, по которому будут определяться выбросы.

    Возвращает:
    ----------
    tuple : (pandas.DataFrame, pandas.DataFrame)
        Первый элемент — DataFrame с потенциальными выбросами.
        Второй элемент — очищенный DataFrame без выбросов.
    """
    # Извлекаем значения признака из DataFrame
    x = data[feature]
    
    # Вычисляем 1-й и 3-й квартили
    quartile_1: float = x.quantile(0.25)
    quartile_3: float = x.quantile(0.75)
    
    # Рассчитываем межквартильный размах (IQR)
    iqr: float = quartile_3 - quartile_1
    
    # Определяем нижнюю и верхнюю границы для выбросов
    lower_bound: float = quartile_1 - (iqr * 0.8)
    upper_bound: float = quartile_3 + (iqr * 0.8)
    
    # Фильтруем данные для получения потенциальных выбросов
    outliers: pd.DataFrame = data[(x < lower_bound) | (x > upper_bound)]
    
    # Создаем очищенный DataFrame, исключая выбросы
    cleaned: pd.DataFrame = data[(x >= lower_bound) & (x <= upper_bound)]
    
    # Возвращаем найденные выбросы и очищенный DataFrame
    return outliers, cleaned
```

### **<center> Метод z-отклонений (метод сигм) </center>** 

Построим две гистограммы признака BMI: первая — в обычном масштабе, а вторая — в логарифмическом. Логарифмировать будем с помощью функции log() из библиотеки numpy (натуральный логарифм — логарифм по основанию числа e). Признак имеет среди своих значений 0. Из математики известно, что логарифма от 0 не существует, поэтому мы прибавляем к нашему признаку 1, чтобы не логарифмировать нули и не получать предупреждения.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 4))

#гистограмма исходного признака
histplot = sns.histplot(combine_data['bmi'], bins=30, ax=axes[0])
histplot.set_title('BMI Distribution');

#гистограмма в логарифмическом масштабе
log_bmi= np.log(combine_data['bmi'] + 1)
histplot = sns.histplot(log_bmi, bins=30, ax=axes[1])
histplot.set_title('Log BMI Distribution');
```

```python
import pandas as pd
import numpy as np

def outliers_z_score(data: pd.DataFrame, feature: str, log_scale: bool = False) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Выявляет выбросы в заданном признаке DataFrame с использованием алгоритма Z-оценки.
    
    Функция принимает на вход DataFrame и название признака, по которому необходимо определить выбросы.
    В дополнение функция может работать в логарифмическом масштабе, если аргумент log_scale установлен в True.
    Функция возвращает два значения: DataFrame с потенциальными выбросами и очищенный DataFrame без них.
    
    Параметры:
    ----------
    data : pandas.DataFrame
        Входной DataFrame, содержащий данные для анализа.
        
    feature : str
        Название признака (колонки) в DataFrame, по которому будут определяться выбросы.
        
    log_scale : bool, optional
        Если True, то применяется логарифмическое преобразование к признаку.
        По умолчанию False.

    Возвращает:
    ----------
    tuple : (pandas.DataFrame, pandas.DataFrame)
        Первый элемент — DataFrame с потенциальными выбросами.
        Второй элемент — очищенный DataFrame без выбросов.
    """
    
    # Если log_scale True, применяем логарифмическое преобразование к признаку
    if log_scale:
        x = np.log(data[feature] + 1)  # Добавляем 1 для предотвращения логарифма от нуля
    else:
        x = data[feature]  # Используем исходные данные

    # Вычисляем среднее и стандартное отклонение
    mu = x.mean()
    sigma = x.std()

    # Определяем нижнюю и верхнюю границы для выбросов
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 3 * sigma

    # Фильтруем данные для получения потенциальных выбросов
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    
    # Создаем очищенный DataFrame, исключая выбросы
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    
    # Возвращаем найденные выбросы и очищенный DataFrame
    return outliers, cleaned
```

Применим эту функцию к таблице combine_data и признаку bmi, а также выведем размерности результатов:

```python
outliers, cleaned = outliers_z_score(combine_data, 'bmi', log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```

### **<center> Обнаружение и ликвидация неинформативных признаков</center>**  

Разберём алгоритм дальнейших действий:

→ Создаём пустой список low_information_cols, куда будем добавлять названия признаков, которые мы посчитаем неинформативными.

→ В цикле пройдёмся по всем именам столбцов в таблице и для каждого будем совершать следующие действия:

- рассчитаем top_freq — наибольшую относительную частоту с помощью метода value_counts() с параметром normalize=True. Метод вернёт долю от общих данных, которую занимает каждое уникальное значение в признаке.

- рассчитаем nunique_ratio — отношение числа уникальных значений в столбце к размеру всего столбца. Число уникальных значений в столбце получим с помощью метода nunique(), а размер признака — с помощью метода count(). Например, для столбца id число уникальных значений — 17495; оно же равно размеру таблицы. Поэтому результат отношения будет 1.
сравним каждое из полученных чисел с пороговым значением (у нас это 0.95) и добавим в список неинформативных признаков, если условие истинно.

```python
#список неинформативных признаков
low_information_cols = [] 

#цикл по всем столбцам
for col in combine_data.columns:
    #наибольшая относительная частота в признаке
    top_freq = combine_data[col].value_counts(normalize=True).max()
    #доля уникальных значений от размера признака
    nunique_ratio = combine_data[col].nunique() / combine_data[col].count()
    # сравниваем наибольшую частоту с порогом
    if top_freq > 0.95:
        low_information_cols.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
    # сравниваем долю уникальных значений с порогом
    if nunique_ratio > 0.95:
        low_information_cols.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
```
